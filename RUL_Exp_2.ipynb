```python
import math
vib_rms = []
square_sum = 0

for i in range(1, m+1):
    square_sum = square_sum + a[i-1]
    rms = math.sqrt(square_sum/i) 
    vib_rms.append(rms)
    print(square_sum, rms)
```

```python
ft = vib_rms.vib_rms.mean() + vib_rms.vib_rms.std()
print(ft)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy.solvers import solve
from sympy import Symbol, exp
from scipy.optimize import curve_fit

n_rows = len(vib_rms)
RUL_trend= []
G_Truth = []

def exponential_func(x, a, b, c):
    return a*np.exp(b*x)+c

for i in range(1, n_rows):
    x = vib_rms.iloc[0:i, 0].values
    y = vib_rms.iloc[0:i, 1].values
    #x = x.reshape(-1,1)
    #print(y[i-1])
    #print(x[-1,0])
    if y[i-1] >=ft: 
        popt, pcov = curve_fit(exponential_func, x, y, p0=[2, 0.5, 1], maxfev = 5000)
        #popt, pcov = curve_fit(exponential_func, x, y, p0=(1, 0,1),maxfev=5000)
        
        popt[0] = round(popt[0],4)
        popt[1] = round(popt[1],4)
        popt[2] = round(popt[2],4)
        
        print(*popt)
        
        z = Symbol('z', positive = True, real = True) 
        remaining = solve(popt[0]*exp(popt[1]*z) +popt[2] - vib_fail , z)
#        print(i)
        print(remaining)
        
```

```python
        RUL = remaining[0] - x[-1]
        actual = n_rows - i
        RUL_trend.append(RUL)
        G_Truth.append(actual)
```

```python
RUL_trend = np.array(RUL_trend)
G_Truth = np.array(G_Truth)
```

```python
df_RUL = pd.DataFrame({'Actual': G_Truth[:],'RUL': RUL_trend[:]})
#df_RUL['diff'] = df_RUL.RUL - df_RUL.Actual
#df_RUL['Error(%)'] = ((df_RUL.Actual - df_RUL.RUL)/df_RUL.Actual)*100
pd.set_option('display.max_rows', df_RUL.shape[0]+1)
df_RUL
```
