```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit


def exponential_func(x, a, b, c):
    return a*np.exp(b*x)+c

x = np.linspace(1,1300,1300)
y = np.array(y[:1300])


popt, pcov = curve_fit(exponential_func, x, y, p0=(1, 1e-6,1), maxfev=2000)

popt[0] = round(popt[0],4)
popt[1] = round(popt[1],4)
popt[2] = round(popt[2],4)

xx = np.linspace(1, 1300, 1300)
yy = exponential_func(xx, *popt)

# please check whether that is correct
r2 = 1. - sum((exponential_func(x, *popt) - y) ** 2) / sum((y - np.mean(y)) ** 2)

plt.plot(x, y, '+', xx, yy)
plt.title('Exponential Fit')
plt.xlabel(r'Time')
plt.ylabel(r'Vib_RMS')
plt.text(650, 1, "equation:\n{:.4f}  exp({:.4f} x) + {:.4f} ".format(*popt))
plt.text(30, 4, "R^2:\n {}".format(r2))
plt.axhline(y = 5.5, color = 'r', linestyle = '-')
plt.show()


print(*popt)


```
